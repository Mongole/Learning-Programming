```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import ast
#Load Data
df = pd.read_csv('C:\\Users\\Joseans-M17\\Documents\\VScode VENV\\data_jobs.csv')
#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#df[['job_title_short', 'job_skills']].head(5)
df_exploded = df.explode('job_skills')
```

```python
df_exploded['job_skills'].value_counts().head(10).plot(kind='bar')
```

```python
skills_count = df_exploded.groupby(['job_title_short','job_skills']).size()
```

```python
df_skills_count = skills_count.reset_index(name='skill_count')

df_skills_count = df_skills_count.sort_values(by='skill_count', ascending=False)

df_skills_count
```

```python
job_title = 'Data Analyst'
top_skills = 10

df_skill_final = df_skills_count[df_skills_count['job_title_short']==job_title].head(top_skills)

df_skill_final.plot(kind='barh', x='job_skills',y='skill_count')
plt.gca().invert_yaxis()
plt.title(f'Top {top_skills} Skills for {job_title}s')
plt.xlabel('Job Posting Count')
plt.ylabel('')
plt.legend().set_visible(False)
plt.show()
```

# Trending Skills excercise

```python
df_DA = df[df['job_title_short']=='Data Analyst'].copy()
df_DA['job_posted_month_no'] = df['job_posted_date'].dt.month
```

```python
df_DA_exploded = df_DA.explode('job_skills')

df_DA_pivot = df_DA_exploded.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

#df_DA_pivot.plot(kind='line')

```

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()

df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_pivot = df_DA_pivot.drop('Total')

df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(kind='line')
plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```
